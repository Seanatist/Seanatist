```python
#linear regression
from sklearn import model_selection, linear_model
import numpy as np
```

```python
x = np.array(dataset["Research"]).reshape(-1,1)
y = np.array(dataset["Total score"]).reshape(-1,1)

#splitting up training/testing sets
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size = 0.2)

#make a linear regression
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)



```

```python
print("Slope/Coefficient:", linear.coef_)
print("Y-Intercept:", linear.intercept_)

#Y = 0.824271x + 10.152185
```

```python
#scoring our model (How good is it?)
print("score", linear.score(x_test, y_test))
```

```python
#Making predictions
predictions = linear.predict(x_test)

for x in range(len(predictions)):
  print("prediction value:", predictions[x], "Actual value:", y_test[x])
```

```python
#Graphing the simple linear regression
plt.scatter(x_train, y_train, color = "black")
plt.plot(x_test, predictions, color = "red", linewidth = 2)
plt.axline(xy1 = (0, float(linear.intercept_)), slope=float(linear.coef_), color = "green")

plt.xlabel("Research")
plt.ylabel("Total score")
plt.title("Total score Vs Research W/ linear Regression")

plt.show()
```
